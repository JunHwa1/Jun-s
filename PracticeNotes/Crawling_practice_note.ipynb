# 실전 2 - 네이버 주가 크롤링
- https://finance.naver.com/sise/sise_quant.nhn
1. 종목명과 현재가를 크롤링해주세요.
2. 전일대비 상승한 종목만 종목명, 현재가, 전일비를 크롤링해주세요

```python
url = 'https://finance.naver.com/sise/sise_quant.nhn'
headers = {'User-Agent':
'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}

response = requests.get(url, headers=headers)
```

```python
response
```

```python
from bs4 import BeautifulSoup as bs 

soup = bs(response.text, 'html.parser')
```

```python
box_tags = soup.select('.type_2>tr')[2:]
```

```python
box_tags
```

```python
result = []

for box_tag in box_tags:
    try:
        temp = {"종목명" : box_tag.select_one('.tltle').text.strip(),
        "현재가" : box_tag.select_one('.number').text.strip()
        }
        result.append(temp)
    except:
        pass    
    
```

```python
result
```

```python
result = []
up_stock = []

for box_tag in box_tags:
    try:
        temp = {"종목명" : box_tag.select_one('.tltle').text.strip(),
        "현재가" : box_tag.select_one('.number').text.strip()
        }
        result.append(temp)
        
        if box_tag.select_one('img'):
            if 'ico_up' in box_tag.select_one('img').attrs['src']:
                up_stock.append({
                    "종목명": box_tag.select_one('.tltle').text.strip(),
                    "현재가": box_tag.select_one('.number').text.strip(),
                    "전일비": box_tag.select('.number')[1].text.strip()
                })
    except:
        pass    
    
```

```python
up_stock
```

## 실전 3 - 무신사 쇼핑몰 크롤링
- 무신사 쇼핑몰 상의(TOP) 카테고리 첫페이지 제품들의 브랜드명, 제품명, 원래가격, 할인가격을 크롤링해주세요.
- https://store.musinsa.com/app/items/lists/001


```python
url = 'https://www.musinsa.com/categories/item/001'
headers = {'User-Agent':
'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
```

```python
response
```

```python
from bs4 import BeautifulSoup as bs 
soup = bs(response.text, 'html.parser')
```

```python
items = soup.select('#searchList .article_info')
print(items)
```

```python
result = []

for item in items:
    
    musinsa = {
        "브랜드명": item.select_one('.item_title').text.strip(),
    "제품명": item.select_one('.list_info').text.strip(),
    "원래가격": item.select_one('.price').text.strip().split('\n')[0],
    "할인가격": item.select_one('.price').text.strip().split('\n')[-1]
    }
    result.append(musinsa)
```

```python
result
```
